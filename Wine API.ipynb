```python
# Trouver tous les scripts sur la page
scripts = soup.find_all('script')

# Parcourir tous les scripts pour trouver celui qui contient les données recherchées
for script in scripts:
    # Utiliser une expression régulière pour trouver le contenu de la variable
    match = re.search(r'window\.__PRELOADED_STATE__\.winePageInformation\s*=\s*({.*?});', script.text)
    if match:
        # Extraire les données JSON de la variable
        wine_information_json = match.group(1)
        
        # Charger les données JSON en tant qu'objet Python
        wine_information = json.loads(wine_information_json)
```

# Scrap du site web le salon des vignerons indépendants

## Scrapping des domaines

```python
# Listes pour stocker les données
names = []
urls = []
stand = []
place = []
```

```python


i = 0
while i <= 200:
    
    urlSVI = f"https://www.vigneron-independant.com/46%C3%A8me-salon-des-vins-des-vignerons-ind%C3%A9pendants-paris?page={i}"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(urlSVI, headers=headers)

    soup = BeautifulSoup(response.text, 'html.parser')

    result_bloks = soup.select('div.result-bloc.node.node-vigneron.search-result--internal')

    if result_bloks == []:
        break
    
    for domaine in result_bloks:
        result_names = domaine.find('div', class_='result-name')
        names.append(result_names.find('a').get_text(strip=True))
        urls.append('https://www.vigneron-independant.com'+result_names.find('a').get('href'))
        
        result_stand = domaine.find('div', class_='result-show-stand')
        stand.append(result_stand.get_text(strip=True))
    
        result_place = domaine.find('div', class_='result-place')
        place.append(result_place.get_text(strip=True))
```

```python
DomaineSVI_Informations = {
    'Name': names,
    'URL': urls,
    'Stand' : stand,
    'Place': place
}

dfDomaine = pd.DataFrame(DomaineSVI_Informations)
```

```python
dfDomaine = dfDomaine.drop_duplicates()
```

```python
result_vin = soup.find_all('div', class_='wine-bloc')

for vin in result_vin:
    names.append(vin.find('div', class_='wine-name').get_text(strip=True))
    place.append(vin.find('div', class_='wine-place').get_text(strip=True))
    domainevendeur.append(dfDomaine['Name'][0])
    domaineproducteururl.append(dfDomaine['URL'][0])
    vinul = vin.find('div', class_='wine-col first')
    li_elements = vinul.find_all('li')
    appellation.append(li_elements[0].get_text(strip=True))
    domaineproducteur.append(li_elements[1].get_text(strip=True))
    c = 0
    t = 0
    ca =0
    ce = 0
    
    for li in li_elements[2:]:
        if li.get_text(strip=True).startswith("Couleur :"):
            couleur.append(li.get_text(strip=True).split(':')[1].strip())
            c = 1
        if li.get_text(strip=True).startswith("Type de produit :"):
            type.append(li.get_text(strip=True).split(':')[1].strip())
            t = 1
        if li.get_text(strip=True).startswith("Caractéristique :"):
            caractéristique.append(li.get_text(strip=True).split(':')[1].strip())
            ca = 1
        if li.get_text(strip=True).startswith("Cépage :"):
            cépage.append(li.get_text(strip=True).split(':')[1].strip())
            ce = 1
    
    if c == 0:
        couleur.append(None)
    if t == 0:
        type.append(None)
    if ca == 0:
        caractéristique.append(None)
    if ce == 0:
        cépage.append(None)
```

```python
VinSVI_Informations = {
    'Name': names,
    'Place': place,
    'Appellation': appellation,
    'Domaine Producteur': domaineproducteur,
    'Domaine Producteur url': domaineproducteururl,
    'Domaine Vendeur': domainevendeur,
    'Couleur': couleur,
    'Type' : type,
    'Caractéristique':caractéristique,
    'Cépage': cépage
}
```

```python
dfVin = pd.DataFrame(VinSVI_Informations)
```

```python


for urlSVIVin in dfDomaine['URL'][500:600]:
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(urlSVIVin, headers=headers)

    soup = BeautifulSoup(response.text, 'html.parser')

    # Listes pour stocker les données vins
    names = []
    place = []
    domainevendeur = []
    domaineproducteur = []
    domaineproducteururl = []
    appellation = []
    couleur = []
    type = []
    caractéristique = []
    cépage = []

    result_vin = soup.find_all('div', class_='wine-bloc')

    for vin in result_vin:
        names.append(vin.find('div', class_='wine-name').get_text(strip=True))
        place.append(vin.find('div', class_='wine-place').get_text(strip=True))
        domainevendeur.append(dfDomaine[dfDomaine['URL'] == urlSVIVin]['Name'].values[0])
        domaineproducteururl.append(urlSVIVin)
        vinul = vin.find('div', class_='wine-col first')
        li_elements = vinul.find_all('li')
        appellation.append(li_elements[0].get_text(strip=True))
        domaineproducteur.append(li_elements[1].get_text(strip=True))
        c = 0
        t = 0
        ca =0
        ce = 0

        for li in li_elements[2:]:
            if li.get_text(strip=True).startswith("Couleur :"):
                couleur.append(li.get_text(strip=True).split(':')[1].strip())
                c = 1
            if li.get_text(strip=True).startswith("Type de produit :"):
                type.append(li.get_text(strip=True).split(':')[1].strip())
                t = 1
            if li.get_text(strip=True).startswith("Caractéristique :"):
                caractéristique.append(li.get_text(strip=True).split(':')[1].strip())
                ca = 1
            if li.get_text(strip=True).startswith("Cépage :"):
                cépage.append(li.get_text(strip=True).split(':')[1].strip())
                ce = 1

        if c == 0:
            couleur.append(None)
        if t == 0:
            type.append(None)
        if ca == 0:
            caractéristique.append(None)
        if ce == 0:
            cépage.append(None)

    VinSVI_Informations = {
        'Name': names,
        'Place': place,
        'Appellation': appellation,
        'Domaine Producteur': domaineproducteur,
        'Domaine Producteur url': domaineproducteururl,
        'Domaine Vendeur': domainevendeur,
        'Couleur': couleur,
        'Type' : type,
        'Caractéristique':caractéristique,
        'Cépage': cépage
    }

    dfVinBoucle = pd.DataFrame(VinSVI_Informations)
    dfVin = pd.concat([dfVin, dfVinBoucle],ignore_index=True)
```
